```python
import pandas as pd
from geopy.distance import geodesic

# 读取CSV文件
file_path = "D:\\研究生\\生成式AI\\wenhui\\9.tucker-知识图谱\\成都市\\成都市station.xlsx"
data = pd.read_excel(file_path)

# 初始化结果列表
triples = []

# 计算每对点之间的距离，并根据距离判断关系
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        name1, lng1, lat1 = data.iloc[i]['name'], data.iloc[i]['lng'], data.iloc[i]['lat']
        name2, lng2, lat2 = data.iloc[j]['name'], data.iloc[j]['lng'], data.iloc[j]['lat']
        
        # 计算经纬度距离
        coord1 = (lat1, lng1)
        coord2 = (lat2, lng2)
        distance = geodesic(coord1, coord2).kilometers
        
        # 判断关系
        if distance < 2:
            triples.append((name1, "borderby", name2))

        elif distance < 5:
            triples.append((name1, "nearby", name2))


# 去重，确保两个站点之间最多只有一种关系
unique_triples = list(set(triples))

for i in range(1, 55):  # 假设有54个站点
    unique_triples.append(("chengdu", "belong", f"station{i}"))
    
# 将三元组存储到txt文件中
output_path = "D:\\研究生\\生成式AI\\wenhui\\9.tucker-知识图谱\\成都市\\knowledge_map_1.txt"
with open(output_path, 'w', encoding='utf-8') as f:
    for triple in unique_triples:
        f.write(f"{triple[0]},{triple[1]},{triple[2]}\n")

print("三元组生成并存储完成。")
```

```python
import pandas as pd
from geopy.distance import geodesic
import os

# 读取换电站数据
station_file_path = r"D:\研究生\生成式AI\wenhui\9.tucker-知识图谱\成都市\成都市station.xlsx"
stations = pd.read_excel(station_file_path)

# 定义POI数据目录
poi_dir = "D:\\研究生\\生成式AI\\wenhui\\1.可用文件\\poi\\成都市"

# 初始化结果列表
results = []

# 计算每个换电站周围2公里半径内的每类POI数量
for i, station in stations.iterrows():
    station_name = station['name']
    station_coord = (station['lat'], station['lng'])
    poi_counts = {'station': station_name}
    
    for poi_file in os.listdir(poi_dir):
        if poi_file.endswith(".csv"):
            poi_category = os.path.splitext(poi_file)[0]
            
            # 尝试使用不同的编码读取CSV文件
            try:
                poi_data = pd.read_csv(os.path.join(poi_dir, poi_file), encoding='utf-8')
            except UnicodeDecodeError:
                try:
                    poi_data = pd.read_csv(os.path.join(poi_dir, poi_file), encoding='gbk')
                except UnicodeDecodeError:
                    poi_data = pd.read_csv(os.path.join(poi_dir, poi_file), encoding='latin1')
            
            poi_data = poi_data.dropna(subset=[poi_data.columns[8], poi_data.columns[9]])
            # 计算每个POI到换电站的距离，并统计在2公里范围内的数量
            count = 0
            for _, poi in poi_data.iterrows():
                poi_coord = (poi.iloc[9], poi.iloc[8])
                #print(poi_coord)
                #print(station_coord)
                distance = geodesic(station_coord, poi_coord).kilometers
                #print(distance)
                if distance <= 2:
                    count += 1
            
            poi_counts[poi_category] = count
    
    results.append(poi_counts)

# 将结果存储到一个CSV文件中
output_file_path = r"D:\研究生\生成式AI\wenhui\9.tucker-知识图谱\成都市\成都市station_POI.csv"
results_df = pd.DataFrame(results)
results_df.to_csv(output_file_path, index=False)

print("POI统计结果已保存到CSV文件。")
```

```python
import re
def split_dataset(file_path):
    train_file_path = r'D:\研究生\生成式AI\wenhui\9.tucker-知识图谱\成都市\knowledge_map_8_train.txt'
    test_file_path = r'D:\研究生\生成式AI\wenhui\9.tucker-知识图谱\成都市\knowledge_map_8_test.txt'
    
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    train_lines = []
    test_lines = []
    
    for line in lines:
        elements = line.strip().split(',')
        if len(elements) == 3:
            first_station_match = re.match(r'station(\d+)', elements[0])
            
            if first_station_match:
                first_station_num = int(first_station_match.group(1))
                
                if 1 <= first_station_num <= 49:
                    train_lines.append(line)
                else:
                    test_lines.append(line)
    
    # 写入训练集文件
    with open(train_file_path, 'w') as train_file:
        for line in train_lines:
            train_file.write(line)
    
    # 写入测试集文件
    with open(test_file_path, 'w') as test_file:
        for line in test_lines:
            test_file.write(line)
    
    print(f'Training set written to {train_file_path}')
    print(f'Test set written to {test_file_path}')

# 使用示例
split_dataset(r'D:\研究生\生成式AI\wenhui\9.tucker-知识图谱\成都市\knowledge_map_8.txt')
```

```python
import pandas as pd
import networkx as nx
from geopy.distance import geodesic

# 读取Excel文件
df = pd.read_excel(r'D:\研究生\生成式AI\wenhui\10.三图卷积\成都市\成都市station.xlsx')

# 创建一个空的无向图
G = nx.Graph()

# 添加节点
for idx, row in df.iterrows():
    G.add_node(row['name'], pos=(row['lng'], row['lat']))

# 初始化距离矩阵
distances = pd.DataFrame(index=df['name'], columns=df['name'])

# 添加边并计算距离
for i, row_i in df.iterrows():
    for j, row_j in df.iterrows():
        station1 = row_i['name']
        station2 = row_j['name']
        if station1 != station2:
            coord1 = (row_i['lat'], row_i['lng'])
            coord2 = (row_j['lat'], row_j['lng'])
            distance = geodesic(coord1, coord2).kilometers
            distances.loc[station1, station2] = round(distance, 4)
            distances.loc[station2, station1] = round(distance, 4)
            if i < j:
                G.add_edge(station1, station2, weight=distance)

# 填充对角线为0（站点到自己的距离为0）
for station in df['name']:
    distances.loc[station, station] = 0.0

# 打印节点和边的信息
print("Nodes:", G.nodes(data=True))
print("Edges:", G.edges(data=True))

# 保存距离矩阵为txt文件
distances.to_csv(r'D:\研究生\生成式AI\wenhui\10.三图卷积\成都市\distances_edge.txt', sep='\t', header=False, index=False, float_format='%.4f')
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, box
from pyproj import Transformer
import math
# 1. 加载 Shapefile 数据
shapefile_path = r'D:\研究生\生成式AI\wenhui\1.可用文件\路网\成都市\成都市.shp'

# 指定目标投影系统 (EPSG 4326 为常见的 WGS84 地理坐标系)
target_crs = {'init': 'epsg:4490'}

try:
    # 尝试使用指定投影系统读取 Shapefile
    gdf = gpd.read_file(shapefile_path).to_crs(target_crs)
except ValueError:
    # 如果失败，使用默认投影系统
    gdf = gpd.read_file(shapefile_path)

# 2. 定义目标位置
station_df = pd.read_excel(r'D:\研究生\生成式AI\wenhui\10.三图卷积\成都市\成都市station.xlsx')

def gcj02_to_wgs84(point):
    x, y = point
    transformer = pyproj.Transformer.from_crs("EPSG:4326", "EPSG:4490", always_xy=True)
    x2, y2 = transformer.transform(x, y)
    return Point(x2, y2)




# 3. 在2公里半径内筛选道路
def filter_roads_within_radius(gdf, target_point, radius_meters=2000):
    # 计算一个近似的度数转换为米的比例
    lat_in_meters = 1 / 111320  # 1度纬度大约等于111.32公里
    lon_in_meters = 1 / (40075000 * math.cos(math.radians(target_point.y)) / 360)  # 经度的度数转换比例根据纬度变化

    # 计算矩形范围
    delta_lat = radius_meters * lat_in_meters
    delta_lon = radius_meters * lon_in_meters
    minx, miny = target_point.x - delta_lon, target_point.y - delta_lat
    maxx, maxy = target_point.x + delta_lon, target_point.y + delta_lat
    bounding_box = box(minx, miny, maxx, maxy)

    # 筛选与矩形范围相交的道路
    filtered_gdf = gdf[gdf.intersects(bounding_box)]


    return filtered_gdf

# 4. 统计道路类型
road_types = ['城市主干道', '城市快速路', '城市次干道', '城市支路', '人行道路', '内部道路', '高速公路', '自行车道', '乡村道路', '其他']
results = []


for _, row in station_df.iterrows():
    name = row['name']
    lat = row['lat']
    lng = row['lng']
    target_point = Point(lng, lat)
    
    # 筛选道路
    filtered_gdf = filter_roads_within_radius(gdf, target_point)
    
    # 统计道路类型
    road_counts = {road_type: filtered_gdf[filtered_gdf['type'] == road_type].shape[0] for road_type in road_types}
    
    # 将结果添加到结果列表
    result = {'name': name}
    result.update(road_counts)
    results.append(result)

# 将结果列表转换为 DataFrame
results_df = pd.DataFrame(results)

# 保存到 CSV 文件
results_df.to_csv(r'D:\研究生\生成式AI\wenhui\10.三图卷积\成都市\成都市station_路网.csv', index=False)

print("结果已保存到 road_statistics.csv 文件中。")

```

```python
import pandas as pd
from sklearn.cluster import KMeans

# 读取Excel文件中的数据
file_path = r'D:\研究生\生成式AI\wenhui\10.三图卷积\成都市\成都市station_路网.csv'
data = pd.read_csv(file_path)

# 提取特征信息（去掉第一列station名字）
features = data.iloc[:, 1:].values

# 使用KMeans进行聚类
num_clusters = 4
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(features)

# 获取聚类结果
cluster_labels = kmeans.labels_

# 获取聚类中心
cluster_centers = kmeans.cluster_centers_

# 寻找每个聚类中心对应的数据点索引和行号
cluster_center_info = []
for center in cluster_centers:
    # 计算每个聚类中心与所有数据点的距离
    distances = np.linalg.norm(features - center, axis=1)
    # 找到最近的数据点索引
    closest_index = np.argmin(distances)
    # 获取对应数据点的行号
    data_point_index = data.index[closest_index]
    cluster_center_info.append(data_point_index)

# 输出聚类结果和对应的数据点行号
print("Cluster labels:")
print(cluster_labels.tolist())
print("\nCluster centers and corresponding data point rows:")
luhan = []
for i, row_index in enumerate(cluster_center_info):
    print(f"Cluster {i+1} center: {cluster_centers[i]}")
    print(f"Corresponding data point row index: {row_index}")
    luhan.append(row_index)
print(luhan)
```

```python
import pandas as pd

# 定义文件路径
csv_file_path = r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\全部.csv'
excel_file_path = r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\成都市\成都市station.xlsx'
output_file_path = r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\成都市\成都市station_换电数据.csv'

# 读取CSV文件
csv_df = pd.read_csv(csv_file_path)

# 读取Excel文件
excel_df = pd.read_excel(excel_file_path)

# 创建一个字典用于code到name的映射
code_to_name = dict(zip(excel_df['code'], excel_df['name']))

# 根据cg_code列映射到name
csv_df['name'] = csv_df['cg_code'].map(code_to_name)

# 筛选出cg_code在code范围内的数据
filtered_df = csv_df[csv_df['cg_code'].isin(excel_df['code'])]

# 输出结果为新的CSV文件
filtered_df.to_csv(output_file_path, index=False)

print("匹配结果已保存到新的CSV文件：", output_file_path)
```

```python
import pandas as pd

# 定义文件路径
input_file_path = r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\成都市\成都市station_换电数据.csv'
output_file_path = r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\成都市\成都市station_换电数据_flow.csv'

# 读取CSV文件
data = pd.read_csv(input_file_path)

data['change_time'] = pd.to_datetime(data['change_time'], format='%d/%m/%Y %H:%M:%S')

data['hour'] = data['change_time'].dt.strftime('%Y-%m-%d %H:00:00')


# 按 name 和 hour 进行分组，并计算每组的条数
grouped_df = data.groupby(['name', 'hour']).size().reset_index(name='flow')

# 按 hour 进行排序
grouped_df = grouped_df.sort_values(by='hour')

# 输出结果为新的CSV文件
grouped_df.to_csv(output_file_path, index=False)

print("分组结果已保存到新的CSV文件：", output_file_path)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# 定义24节气
solar_terms = {
    "立春": "2024-02-04", "雨水": "2024-02-19", "惊蛰": "2024-03-05", "春分": "2024-03-20",
    "清明": "2024-04-04", "谷雨": "2024-04-20", "立夏": "2024-05-05", "小满": "2024-05-21",
    "芒种": "2024-06-05", "夏至": "2024-06-21", "小暑": "2024-07-06", "大暑": "2024-07-22",
    "立秋": "2024-08-07", "处暑": "2024-08-23", "白露": "2024-09-07", "秋分": "2024-09-23",
    "寒露": "2024-10-08", "霜降": "2024-10-23", "立冬": "2024-11-07", "小雪": "2024-11-22",
    "大雪": "2024-12-07", "冬至": "2024-12-21", "小寒": "2025-01-05", "大寒": "2025-01-20"
}

# 节假日
holidays = [
    "2024-02-10", "2024-02-11", "2024-02-12", "2024-02-13", "2024-02-14", "2024-02-15", "2024-02-16", "2024-02-17",
    "2024-04-04", "2024-04-05", "2024-04-06",
    "2024-05-01", "2024-05-02", "2024-05-03", "2024-05-04", "2024-05-05",
    "2024-06-08", "2024-06-09", "2024-06-10"
]

# 转换字符串为日期
solar_terms = {k: datetime.strptime(v, "%Y-%m-%d") for k, v in solar_terms.items()}
holidays = [datetime.strptime(date, "%Y-%m-%d") for date in holidays]

# 生成特征数据
def generate_features(start_date, end_date):
    data = []
    current_date = start_date

    while current_date <= end_date:
        hour = current_date.hour
        week = current_date.weekday() + 1
        is_holiday = 1 if current_date in holidays or current_date.weekday() >= 5 else 0

        season = 0
        for i, (term, date) in enumerate(solar_terms.items()):
            if current_date < date:
                season = i
                break

        data.append({
            "datetime": current_date,
            "hour": hour,
            "week": week,
            "season": season,
            "holiday": is_holiday
        })

        current_date += timedelta(hours=1)

    df = pd.DataFrame(data)
    return df

# 设置开始和结束日期
start_date = datetime(2024, 1, 1, 0, 0, 0)
end_date = datetime(2024, 6, 28, 0, 0, 0)

# 生成特征数据
df_features = generate_features(start_date, end_date)


# 查看生成的特征数据并保存为CSV文件
df_features.to_csv(r'D:\研究生\生成式AI\wenhui\11.transformer时间特征\成都市\condition_2024.csv', index=False)

```
